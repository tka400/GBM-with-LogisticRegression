# Gradient Boosting Machine with Logistic Regression as an base algorithm

## Data

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

X, y = datasets.make_classification(n_samples=2000, n_features=10, n_informative=5)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## Algorithm`s classes

```python
class GBM:
    def __init__(self, n_estimators=50, lr=0.1):
        self.n_estimators = n_estimators
        self.trees = []
        self.residuals = []
        self.first_leaf = None
        self.lr = 0.1
        self.train_score = []
        
    def fit(self, X, y):
        self.first_leaf = y.mean()
        prediction = np.ones(len(y)) * self.first_leaf
        
        for i in range(self.n_estimators):
            residuals = y - prediction
            self.residuals.append(residuals)
            
            tree = LogisticRegression()
            tree.fit(X, residuals)
            self.trees.append(tree)
            self.train_score.append(self.score(self.predict(X, len(self.trees)), y))
            
            prediction += self.lr * tree.predict(X)
            
    def predict(self, X, n):
        predictions = np.ones(X.shape[0]) * self.first_leaf
        
        for i in range(n):
            predictions += self.lr * self.trees[i].predict(X)
        
        threshold = 0.5
        labels = [0 if i <= threshold else 1 for i in predictions]
        
        return labels
    
    def score(self, predicted, y):
        return np.sum(predicted == y) / len(y)
    
    def roc(self, X, y, n):
        predictions = np.ones(X.shape[0]) * self.first_leaf
        
        for i in range(n):
            predictions += self.lr * self.trees[i].predict(X)
            
        indexes = np.argsort(predictions)    
        
        TPR, FPR, P, R = [], [], [], []
        
        for threshold in predictions[indexes]:
            labels = [0 if i <= threshold else 1 for i in predictions]
            
            TP, FP, FN, TN = 0, 0, 0, 0
            
            for i in range(len(labels)):
                if labels[i] == 0 and y[i] == 0: TP += 1
                if labels[i] == 1 and y[i] == 1: TN += 1
                if labels[i] == 0 and y[i] == 1: FP += 1
                if labels[i] == 1 and y[i] == 0: FN += 1                        
            
            TPR.append(TP / (TP + FN))
            FPR.append(FP / (FP + TN))
            R.append(TP / (TP + FN))
            P.append(TP / (TP + FP))
            
        return TPR, FPR, P, R
```

```python
class LogisticRegression:
    def __init__(self, iters=200, learning_rate=0.1):
        self.iters = iters
        self.lr = learning_rate
        self.weights = None
        
    def fit(self, X, y):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)       
        self.weights = np.ones(X.shape[1])

        for i in range(self.iters):
            predictions = X.dot(self.weights)
            y_predicted = 1 / (1 + np.exp(-predictions))
            error = y_predicted - y
            
            self.dw = 2 / X.shape[0] * X.T.dot(error)
            self.weights -= self.lr * self.dw         
    
    def predict(self, X):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        predictions = X.dot(self.weights)
        y_predicted = 1 / (1 + np.exp(-predictions))
        return y_predicted
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
predict = clf.predict(X_test)
```

```python
np.sum(np.where(predict < 0.5, 0, 1) == y_test) / len(y_test)
```

## Model

```python
reg = GBM(n_estimators=100, lr=0.1)
reg.fit(X_train, y_train)
```

```python
plt.subplots(figsize=(4, 4))
plt.plot(reg.residuals)
plt.title("residuals vs estimators amount")
plt.xlabel("n estimators")
plt.ylabel("residuals")

plt.savefig("residuals.png")
plt.show()
```

## Metrics

```python
test_score = []

for i in range(len(reg.trees)):
    predict = reg.predict(X_test, i)
    test_score.append(reg.score(predict, y_test))
```

```python
plt.subplots(figsize=(4, 4))
plt.plot(1 - np.array(reg.train_score), label="train error")
plt.plot(1 - np.array(test_score), label="test error")
plt.legend()
plt.ylabel("error")
plt.xlabel("n estimators")

plt.savefig("errors.png")
plt.show()
```

```python
TPR, FPR, P, R = reg.roc(X_test, y_test, len(reg.trees))
```

```python
plt.subplots(figsize=(4, 4))
plt.plot(FPR, TPR, label="ROC")
plt.plot(R, P, label="PRC")

plt.legend()
plt.title("ROC and PRC")

plt.savefig("curves.png")
plt.show()
```

## Model quality by Poisson Bootstrap

```python
N = len(y_test)
B = 100
```

```python
P, R = [], []

for i in tqdm(range(B)):
    bootstraps = np.random.poisson(1, N)
    TP, FP, TN, FN = 0, 0, 0, 0
    
    for j in range(N):
        if predict[j] == 0 and y_test[j] == 0: TP += bootstraps[j]
        if predict[j] == 1 and y_test[j] == 1: TN += bootstraps[j]
        if predict[j] == 1 and y_test[j] == 0: FN += bootstraps[j]
        if predict[j] == 0 and y_test[j] == 1: FP += bootstraps[j]
        
    P.append(TP / (TP + FP))
    R.append(TP / (TP + FN))
```

```python
plt.hist(R, label="recall", bins=20)
plt.hist(P, label="precision", bins=20)
plt.legend()
plt.xlim(0.0, 1.05)

plt.xlabel("score")
plt.savefig("bootstrap.png")
plt.show()
```
